```python
driver=webdriver.Chrome()
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
driver.execute_script("window.scrollTo(0, 1600);")
```

```python
alldetails=[]
ald=driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td')
for a in ald:
    alldetails.append(a.text)
    
rank=alldetails[0::6]
name=alldetails[1::6]
artist=alldetails[2::6]
uploadDate=alldetails[4::6]
views=alldetails[3::6]
```

Scrape the details team India’s international fixtures from bcci.tv.
Url = https://www.bcci.tv/.
You need to find following details:

    A) Match title (I.e. 1 st ODI)
    B) Series
    C) Place
    D) Date
    E) Time
    Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
driver=webdriver.Chrome()
driver.get('https://www.bcci.tv/')
time.sleep(2)

international=driver.find_element_by_xpath('//div[@class="navigation__drop-down drop-down drop-down--reveal-on-hover"]')
international.click()
time.sleep(2)

fixture=driver.find_element_by_xpath('/html/body/div[3]/div/div[2]/div[2]/nav/ul/li[1]/div[2]/div/ul/li[1]/a')
fixture.click()
time.sleep(2)

matchtitle=[] # match title
matchtitlE=driver.find_elements_by_xpath('//div[@class="fixture__format-strip"]')
for mt in matchtitlE:
    matchtitle.append(mt.text.replace('\n',' '))

#time.sleep(2)

#series and place
seriesandplace=[]
seriesnplacE=driver.find_elements_by_xpath('//p[@class="fixture__additional-info"]')
for p in seriesnplacE:
    seriesandplace.append(p.text.split('\n'))
    
series=[s[0] for s in seriesandplace] # series
place=[p[1] for p in seriesandplace ] # place

# retrive date and time
dateandtime=[]
datentime=driver.find_elements_by_xpath('//div[@class="fixture__full-date"]')
for dnt in datentime:
    dateandtime.append(dnt.text.split('\n'))
    
date=[d[0]+' '+d[1] for d in dateandtime]
time=[t[2] for t in dateandtime]
```

```python
bcci=pd.DataFrame()
bcci['Match Title']=matchtitle
bcci['Series']=series
bcci['Place']=place
bcci['Date']=date
bcci['Time']=time

bcci
```

Scrape the details of Data science recruiters from naukri.com.
Url = https://www.naukri.com/
You have to find the following details:

    A) Name
    B) Designation
    C) Company
    D) Skills they hire for
    E) Location
Note: - From naukri.com homepage click on the recruiters option and the on the search pane type Data science and
click on search. All this should be done through code

```python
driver=webdriver.Chrome()
driver.get('https://www.naukri.com/')
recruiter=driver.find_element_by_xpath('//a[@title="Search Recruiters"]')
url=recruiter.get_attribute('href')
driver.get(url)

search=driver.find_element_by_xpath('//input[@rel="custom:1016"]')
search.send_keys('Data Science')
search.send_keys(Keys.ENTER)
```

```python
count=1
person=[]
while(count<=119):
    try:
        if count==1:
            
            personprofile=driver.find_elements_by_xpath('//a[@class="ellipsis"]')
            for pr in personprofile:
                person.append(pr.get_attribute('href'))
    
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            nextbutton=driver.find_element_by_xpath('//button[@class="grayBtn"]')
            nextbutton.click()
        else:
            personprofile=driver.find_elements_by_xpath('//a[@class="ellipsis"]')
            for pr in personprofile:
                person.append(pr.get_attribute('href'))
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            nextbutton=driver.find_element_by_xpath('/html/body/div[3]/div/div[2]/div/div[3]/button[2]')
            nextbutton.click()
    except:
        pass
    count+=1
```

```python
namE=[]
designation=[]
company=[]
location=[]
skill=[]

for i in person:
    driver.get(i)
    try:
        name=driver.find_element_by_xpath('/html/body/div[2]/div[1]/div[1]/div[2]/div[2]/h1')
        namE.append(name.text)
    except NoSuchElementException:
        pass
    try:
        
        designatioN=driver.find_element_by_xpath('//div[@class="ellipsis"]')
        designation.append(designatioN.text)
    
    except NoSuchElementException:
        pass
    
    try:
        
        companY=driver.find_element_by_xpath('/html/body/div[2]/div[1]/div[1]/div[2]/div[4]/a')
        company.append(companY.text)
    except NoSuchElementException:
        pass
    
    try:
    
        locatioN=driver.find_element_by_xpath('/html/body/div[2]/div[1]/div[1]/div[2]/div[5]/a')
        location.append(locatioN.text)
    except NoSuchElementException:
        pass
    
    try:
            
        skilL=driver.find_element_by_xpath('/html/body/div[2]/div[1]/div[2]/div[2]/div[1]/p[1]')
        skill.append(skilL.text)
    except NoSuchElementException:
        pass
    if len(namE)==300: # here I  bound the limit of execution 
        break
```

```python
recruiter=pd.DataFrame()
recruiter['Name']=namE
recruiter['Designation']=designation
recruiter['Company']=company
recruiter['Skill']=skill
recruiter['Location']=location

recruiter
```

Scrape the details most watched tv series of all time from imdb.com.
Url = https://www.imdb.com/list/ls095964455/
You have to find the following details:
    
    A) Name
    B) Year span
    C) Genre
    D) Run time
    E) Ratings
    F) Votes

```python
driver=webdriver.Chrome()
driver.get('https://www.imdb.com/list/ls095964455/')
names=[]
name=driver.find_elements_by_xpath('//h3[@class="lister-item-header"]//a')
for n in name:
    names.append(n.text)
years=[]
year=driver.find_elements_by_xpath('//span[@class="lister-item-year text-muted unbold"]')
for y in year:
    years.append(y.text)
genera=[]
generA=driver.find_elements_by_xpath('//span[@class="genre"]')
for g in generA:
    genera.append(g.text)
    
runtime=[]
runtimE=driver.find_elements_by_xpath('//span[@class="runtime"]')
for r in runtimE:
    runtime.append(r.text)
    
ratings=[]
rating=driver.find_elements_by_xpath('//div[@class="ipl-rating-star small"]')
for ra in rating:
    ratings.append(ra.text)
ratings

otherdata=[]
other=driver.find_elements_by_xpath('//p[@class="text-muted text-small"]')
for o in other:
    otherdata.append(o.text)
    
votes=otherdata[2::3]
votes=[v.replace('Votes: ','') for v in votes]

imdb=pd.DataFrame()
imdb['Name']=names
imdb['Year']=years
imdb['Genera']=genera
imdb['Runtime']=runtime
imdb['Rating']=ratings
imdb['Vote']=votes

imdb
```
