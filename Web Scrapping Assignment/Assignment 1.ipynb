## Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests
import numpy as np
import pandas as pd
# import warnings
# warnings.filterwarnings("ignore", category=UserWarning, module='bs4')
```

```python
def wikipediaheader(url):
    html=urlopen(url)
    soup=BeautifulSoup(html,"html.parser")
    titles=soup.find_all(['h1','h2','h3','h4','h5','h6'])
    print('All Header tags :',*titles,sep='\n\n')
    
url='https://en.wikipedia.org/wiki/Main_Page'
wikipediaheader(url)
```

## Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and make data frame.

```python
def imdbtop100(url):
    name=[]
    year=[]
    rating=[]
    url_text=requests.get(url).text
    soup=BeautifulSoup(url_text,'html.parser')
    movies=soup.findAll('td',{'class':'titleColumn'})
    moviesrating=soup.findAll('td',{'class':'ratingColumn imdbRating'})
    for i in movies[1:101]:
        name.append(i.a.text)
        year.append(i.span.text[1:5])
    for j in moviesrating[1:101]:
        rating.append(j.text[1:4])
        
    imdb100=pd.DataFrame()
    imdb100['Movies Name']=name
    imdb100['Year']=year
    imdb100['Rating']=rating
    imdb100.index = np.arange(1, len(imdb100)+1) # start index from 1
    return imdb100

url='https://www.imdb.com/chart/top/?ref_=nv_mv_250'
imdbtop100(url)
```

### Write a python program to scrap book name, author name, genre and book review of any 5 books from
‘www.bookpage.com’

```python
def bookpage(url):
    bookname=[]
    author=[]
    genre=[]
    reviews=[]
    urls=[]
    newurls=[]
    baseurl='https://bookpage.com/'
    url_text=requests.get(url).text
    soup=BeautifulSoup(url_text,'html.parser')
    booknamedata=soup.findAll('div',{'class':'bp-block article-info'})
    
    for data in booknamedata[0:5]:
        bookname.append(data.h4.text.replace('\n',''))
        author.append(data.p.text.replace('\n',''))
        genre.append(data.find('p',class_='genre-links hidden-phone').text.replace('\n',''))
        
        
    #for fatching review
    url_tags=soup.find_all("h4",class_="italic")  
    for i in url_tags:
        for j in i.find_all('a',href=True):
            if j.text:
                urls.append(j['href'])
                
    for n in urls:
        newurls.append(baseurl+n)
    for m in newurls:
        req=requests.get(m)
        soup_=BeautifulSoup(req.content,'html.parser')
        review=soup_.find('div',class_='article-body').text.replace('\n','')
        reviews.append(review)

        
    book=pd.DataFrame()
    book['Book Name']=bookname
    book['Author']=author
    book['Genre']=genre
    book['Review'] = reviews[:5]
    book.index=np.arange(1,len(book)+1)
    return book

url='https://bookpage.com/reviews'
bookpage(url)
```

### Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have to scrape:
i) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.  
ii) Top 10 ODI Batsmen in men along with the records of their team and rating.  
iii) Top 10 ODI bowlers along with the records of their team and rating.

```python
def top10odiMenTeam(url):
    request=requests.get(url)
    soup=BeautifulSoup(request.content,'html.parser')
    
    table=soup.find('table',class_='table')
    matchdata=[]
    tabledata=table.tbody.findAll('td')
    for data in tabledata:
        matchdata.append(data.text.strip())
        
    team=matchdata[1::5]
    teams=[]
    for t in team:
        tm=t.split('\n')
        teams.append(tm[0])
    matches=matchdata[2::5]
    points=matchdata[3::5]
    ratings=matchdata[4::5]
    
    top10teams=pd.DataFrame()
    top10teams['Team']=teams[:10]
    top10teams['Matches']=matches[:10]
    top10teams['Points']=points[:10]
    top10teams['Ratings']=ratings[:10]
    top10teams.index=np.arange(1,len(top10teams)+1)
    
    return top10teams    

page='https://www.icc-cricket.com/rankings/mens/team-rankings/odi'
top10odiMenTeam(page)
```

```python
def mentop10batingODI(url):
    request=requests.get(url)
    soup=BeautifulSoup(request.content,'html.parser')
    table=soup.find('table',class_='table rankings-table')
    tabledata=table.findAll('td')
    data=[]
    for d in tabledata:
        data.append(d.text.strip())
    playername=data[1::5]
    team=data[2::5]
    rating=data[3::5]
    
    df=pd.DataFrame()
    df['Name']=playername[:10]
    df['Team']=team[:10]
    df['Rating']=rating[:10]
    df.index=np.arange(1,len(df)+1)
    
    return df
page='https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting'
mentop10batingODI(page)
```

```python
def mentop10bowling(url):
    request=requests.get(url)
    soup=BeautifulSoup(request.content,'html.parser')
    bowlingtable=soup.find('table',attrs={'class':'table rankings-table'})
    table=bowlingtable.findAll('td')
    tabledata=[]
    for data in table:
        tabledata.append(data.text.replace('\n',' ').replace('Pos Player Team Rating Career Best Rating','').strip())
    tabledata.pop(0)
    name=tabledata[::5]
    team=tabledata[1::5]
    rating=tabledata[2::5]
    
    df=pd.DataFrame()
    df['Name']=name[:10]
    df['Team']=team[:10]
    df['Rating']=rating[:10]
    df.index=np.arange(1,len(df)+1)
    return df

page='https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling'
mentop10bowling(page)
```

### Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title, company name, CTC, and apply date.

```python
def internsalafresherjob(url):
    jobTilte=[]
    companyName=[]
    req=requests.get(url)
    soup=BeautifulSoup(req.content)
    
    jobtitle=soup.find_all('div',class_="heading_4_5 profile")    
    for i in jobtitle:
        jobTilte.append(i.text.replace('\n','').strip())
        
    companyname=soup.find_all('a',class_='link_display_like_text')
    for company in companyname:
        companyName.append(company.text.strip())
        
    otherdetails=[]
    otherdetail=soup.find_all('div',class_='item_body') # ctc and apply date
    for detail in otherdetail:
        otherdetails.append(detail.text.strip().replace('\n',''))
    ctc=otherdetails[1::3]
    applydate=otherdetails[2::3]
    
    data=pd.DataFrame()
    data['Job title']=jobTilte
    data['Company Name']=companyName
    data['CTC']=ctc
    data['Apply date']=applydate
    data.index=np.arange(1,len(data)+1)
    return data

page='https://internshala.com/fresher-jobs'
internsalafresherjob(page)
```

### Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The scraped data should include Product Name, Price, Image URL and Average Rating.

```python
def mobileunder20000(url):
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
    request=requests.get(url,headers=headers).text
    soup=BeautifulSoup(request,'lxml')
    
    productname=[]
    for product in soup.findAll('span',class_='a-size-medium a-color-base a-text-normal'):
        productname.append(product.text)
        
    prices=[]
    for price in soup.findAll('span',class_='a-price-whole'):
        prices.append(price.text)
        
    imageurl=[]
    for link in soup.find_all('img',class_='s-image'):
        imageurl.append(link.get('src'))
        
    ratings=[]
    for rating in soup.find_all('span',class_='a-icon-alt'):
        ratings.append(rating.text)
        
    mobileunder20k=pd.DataFrame()
    mobileunder20k['Product']=productname
    mobileunder20k['Price']=prices[:len(productname)]
    mobileunder20k['Image url']=imageurl
    mobileunder20k['Rating']=ratings[:len(imageurl)]
    mobileunder20k.index=np.arange(1,len(mobileunder20k)+1)
    
    return mobileunder20k

amazon_url='https://www.amazon.in/s?k=mobile+under+20000&ref=nb_sb_noss_1'
mobileunder20000(amazon_url)
```
