#### Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.

```python
driver=webdriver.Chrome()
url='https://www.naukri.com/'
driver.get(url)
time.sleep(2)
searchJob=driver.find_element_by_id('qsb-keyword-sugg')
time.sleep(2)
searchJob.send_keys('Data Analyst')
searchJoblocation=driver.find_element_by_id('qsb-location-sugg')
searchJoblocation.send_keys('Bangalore')
time.sleep(2)
search=driver.find_element_by_class_name('search-btn')
search.click()
time.sleep(2)
jobtitle=[]
otherdetail=[]

job_titles=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
for title in job_titles:
    jobtitle.append(title.text)
    
otherdetails=driver.find_elements_by_xpath('//span [@class="ellipsis fleft fs12 lh16"]')
for details in otherdetails:
    otherdetail.append(details.text)
    

experiance=otherdetail[0::3]
location=otherdetail[2::3]

company_name=[]
companyname=driver.find_elements_by_xpath('//a[@class="subTitle ellipsis fleft"]')
for i in companyname:
    company_name.append(i.text)
    
df=pd.DataFrame()
df['Job title']=jobtitle
df['Location']=location
df['Experiance']=experiance
df['Company Name']=company_name

df
```

### Write a python program to scrape data for “Data Scientist” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, full job-description. You have to scrape first 10 jobs data.

```python
driver=webdriver.Chrome()
url='https://www.naukri.com/'
driver.get(url)
time.sleep(2)
searchJob=driver.find_element_by_id('qsb-keyword-sugg')
time.sleep(2)
searchJob.send_keys('Data Analyst')
searchJoblocation=driver.find_element_by_id('qsb-location-sugg')
searchJoblocation.send_keys('Bangalore')
time.sleep(2)
search=driver.find_element_by_class_name('search-btn')
search.click()
time.sleep(2)
jobtitle=[]
otherdetail=[]

job_titles=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
for title in job_titles:
    jobtitle.append(title.text)
    
otherdetails=driver.find_elements_by_xpath('//span [@class="ellipsis fleft fs12 lh16"]')
for details in otherdetails:
    otherdetail.append(details.text)
    
location=otherdetail[2::3]

company_name=[]
companyname=driver.find_elements_by_xpath('//a[@class="subTitle ellipsis fleft"]')
for i in companyname:
    company_name.append(i.text)
    
urls=[]
for url in driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]'):
    urls.append(url.get_attribute('href'))
    
descriptions=[]
for link in urls[:10]:
    driver.get(link)
    time.sleep(3)
    #fetching job description
    try:
        description=driver.find_element_by_xpath('//div[@class="dang-inner-html"]')
        descriptions.append(description.text.replace('\n',''))
    except NoSuchElementException:
        descriptions.append('No Job description Available')
        
df2=pd.DataFrame()
df2['Job title']=jobtitle[:10]
df2['Location']=location[:10]
df2['Company']=company_name[:10]
df2['Job description']=descriptions

df2
```

### Scrape data of first 100 sunglasses listings on flipkart.com. 
You have to scrape four attributes:
1. Brand
2. Product Description
3. Price
4. Discount %

```python
driver=webdriver.Chrome()
url="https://www.flipkart.com/"
driver.get(url)
time.sleep(2)

search=driver.find_element_by_xpath('//input[@type="text"]')
search.send_keys('Sunglasses')
time.sleep(2)
searchicon=driver.find_element_by_xpath('//button[@type="submit"]')
searchicon.click()
time.sleep(2)

urls=[]
for url in driver.find_elements_by_xpath('//a[@class="IRpwTa"]'):
    urls.append(url.get_attribute('href'))
brands=[]
productdescriptions=[]
prices=[]
discounts=[]

for urls_ in urls:
    driver.get(urls_)
    # fetch product brand
    brand=driver.find_elements_by_xpath('//span[@class="G6XhRU"]')
    for b in brand:
        brands.append(b.text)
    # product description 
    productdescription= driver.find_elements_by_xpath('//span[@class="B_NuCI"]')
    for pd in productdescription:
        productdescriptions.append(pd.text)
        
    # price
    price=driver.find_elements_by_xpath('//div[@class="_30jeq3 _16Jk6d"]')
    for p in price:
        prices.append(p.text)
        
    # discounts
    discount=driver.find_elements_by_xpath('//div[@class="_3Ay6Sb _31Dcoz pZkvcx"]/span')
    for d in discount:
        discounts.append(d.text)

import pandas as pd        
sunglasses=pd.DataFrame()
sunglasses['Brand']=brands
sunglasses['Product Description']=productdescriptions
sunglasses['Price']=prices
sunglasses['Discount']=discounts

sunglasses
```
